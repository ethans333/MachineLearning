```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from word2number import w2n
```

```python
df = pd.read_csv("./Data Files/hiring.csv")
df
```

```python
df.experience = df.experience.fillna('zero')
df.experience = df.experience.apply(w2n.word_to_num)
df
```

```python
median_test_score = df['test_score(out of 10)'].median()
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(median_test_score)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']], df['salary($)'])
```

```python
df = pd.DataFrame([[2, 9, 6], [12, 10, 10]], columns=['experience', 'test_score(out of 10)', 'interview_score(out of 10)'])

df.insert(3, "Predicted Salary", reg.predict([[2, 9, 6], [12, 10, 10]]))
df
```
